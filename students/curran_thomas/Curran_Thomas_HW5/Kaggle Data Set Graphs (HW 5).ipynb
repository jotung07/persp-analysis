```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv("MA_Public_schools_2017.csv")
```

```python
%matplotlib inline
df2 = df[["School Name",
          "District Name",
          "First Language Not English",
          "Students With Disabilities",
          "High Needs",	
          "Economically Disadvantaged",
          "TOTAL_Enrollment"]]

df2 = df2.groupby("District Name").sum()

df2["Log Total Enrollment"] = df2['TOTAL_Enrollment'].apply(np.log)

df2["Percent FRL"] = round(df2["Economically Disadvantaged"] / df2["TOTAL_Enrollment"],2)*100
df2["Percent First Language Not English"] = round(df2["First Language Not English"] / df2["TOTAL_Enrollment"],2)*100
df2["Percent Students With Disabilities"] = round(df2["Students With Disabilities"] / df2["TOTAL_Enrollment"],2)*100
df2["Percent High Needs"] = round(df2["High Needs"] / df2["TOTAL_Enrollment"],2)*100

df2.head()

```

```python
#df2.plot.scatter("Percent FRL", "Percent Students With Disabilities")

sns.lmplot(x="Percent FRL", 
           y="Percent Students With Disabilities", 
           data = df2, 
           fit_reg = True)

plt.ylim(0,max(df2["Percent Students With Disabilities"]))
plt.xlim(0,100)


```
