```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
schools = pd.read_csv("MA_Public_Schools_2017.csv")
```

```python
%matplotlib inline

frl_dropout_plot = schools.plot.scatter("% Economically Disadvantaged", "% Dropped Out")
frl_dropout_plot.set_yticklabels(['{:3.2f}%'.format(x) for x in frl_dropout_plot.get_yticks()] )
frl_dropout_plot.set_xticklabels(['{:3.2f}%'.format(x) for x in frl_dropout_plot.get_yticks()] )
plt.title("Economically Disadvantaged Students and Dropout Rates")
frl_dropout_plot.figure.savefig("frl_dropout_plot.pdf")
frl_dropout_plot.figure.savefig("frl_dropout_plot.png")
```
