# Data Preperation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
raw = pd.read_csv('gss2012.csv')
mask = raw['income06'].notnull()
df = raw[mask]
income_dict = {'$150000 OR OVER':15, '$110000 TO $129999':10, '$130000 TO $149999':13,
       '$50000 TO 59999':5, '$25000 TO 29999':2.5, '$30000 TO 34999':3.0,
       '$5 000 TO 5 999':0.5, 'UNDER $1 000':0.0, '$8 000 TO 9 999':0.8,
       '$22500 TO 24999':2.25, '$40000 TO 49999':4, '$60000 TO 74999':6,
       '$90000 TO $109999':9, '$75000 TO $89999':7.5, '$35000 TO 39999':3.5,
       '$15000 TO 17499':1.5, '$20000 TO 22499':2, '$6 000 TO 6 999':0.6,
       '$17500 TO 19999':1.75, '$12500 TO 14999':1.25, '$10000 TO 12499':1,
       '$1 000 TO 2 999':0.1, '$7 000 TO 7 999':0.7, '$4 000 TO 4 999':0.4,
       '$3 000 TO 3 999':0.3}
df['income'] = df['income06'].apply(lambda x: income_dict[x])
```

## Return to Education
In the last part, I want to talk something about the reward of the education. From figure 6, we could see that people does not finish high school has less than 20 000 us dollar median annual income. For people who finished the high school, the median is almost doubled to reach nearly 40 000 us dollar. For people finished junior college, the median income does not change much, but the upper 75% quantile income of the group improved form 60 000 us dollar to 80 000 us dollar. For those finished the bachelor degree, the median income is doubled again. For graduate degree, the payoff is not small. It improves from less than 80 000 to 90 000 us dollar, and more significantly the higher 75% quantile improve from 100 000 to 150 000, a 50% percent of increase. The outlier are also very interesting, for people who does not finish high school or just have high school diploma are also possible to reach higher income quantile, thought that is just a outlier in the distribution as we could see from the box plot in figure 6.

This statistical does provide a correlation of education and income, but it does not imply that for a given person and then receives more education can earn more. Because it is possible that this result is just purely a self-selection bias. People who are either have higher social economical status or have a stronger ability could obtain more education, then it appears that people with more education has higher income. However, this does not reflect the person are make money because they receive education, they might earn the same if they does not receive those eudcation. We might be interested in further exploring on how education and income covariate.

```python
sns.countplot(x='income', data=df)
plt.xlabel('Density')
plt.ylabel('Individual Income (in tens of thousands dollar)')
plt.title('Fig1: Income Bar Plot of GSS Survey Data')
plt.show()
```

```python
sns.violinplot(x='sex', y='income', data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Fig2: Income Violin Plot by Gender')
plt.show()
```

```python
sns.violinplot(x='race', y='income', data=df)
plt.xlabel('Race')
plt.ylabel('Individual Income (in tens of thousands dollar)')
plt.title('Fig3: Income Box Plot by Race')
plt.show()
```

```python
sns.violinplot(y='income', x='voted08', data=df)
plt.xlabel('Count')
plt.ylabel('Individual Income (in tens of thousands dollar)')
plt.title('Fig4: Income Box Plot by 2008 Voting')
plt.show()
```

```python
my_sticks = ['Independent', 'Repulican', 'Democrats']
sns.violinplot(x='partyid_3', y='income', data=df)
plt.xlabel('Party Affliation')
plt.ylabel('Individual Income (in tens of thousands dollar)')
plt.title('Fig4: Income Box Plot by Different Party Affliation')
plt.xticks(np.array([0,1,2]), my_sticks)
plt.show()
```

```python
degree_list = ['<HS', 'HS', 'Junior Coll', 'Bachelor deg','Graduate deg']
my_sticks = ['Not Finish High School', 'High School', 'Junior College', 'Bachelor','Graduate']
sns.boxplot(x='degree', y='income',  order=degree_list, data=df)
plt.xlabel('Highest Degree')
plt.ylabel('Individual Income (in tens of thousands dollar)')
plt.title('Fig5: Income Box Plot by Education Level')
plt.xticks(np.array([0,1,2,3,4]), my_sticks)
plt.show()
```
