I want to describe the "Spooky Author Identification" competition in Kaggle. The goal of this competition is to identify the author of the randomly chosen texts from three authors. The competition randomly draws a sample of text from a database of the horror stories written by Edgar Allan Poe, Mary Shelley, and HP Lovecraft. The participants are asked to predict what are the probabilities of the certain texts is written by three above mentioned authors. Submissions are evaluated using multi-class logarithmic loss. Each text has a unique id, and it connect with the name of the true author. For each id, you must submit a predicted probabilities for each author. The evaluation formula is given in the following:

$$logloss = - \frac{1}{N} \sum_{i=1}^{N}  \sum_{j=1}^{M} y_{ij} log(p_{ij})$$

where N is the number of observation in the test set, M is the number of class labels, $y_{ij}$ is 1 if observation i belongs to class j and 0 otherwise, and $p_{ij}$ is the predicted probability the observation i belongs to class j.

### Problem 3

The is the dataset of median income of US Cities. In the following, I will draw a plot of the fitted the distribution of median income of over the more than 25,000 cities in US. We could see a large inequality across USA cities. There are cities very poor has lower than 20,000 of median household income, and with extremly wealthy cities with 250,000 of median household income that can be barely seen from the vilion plot.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from scipy import stats

income = pd.read_csv('MedianHouseholdIncome2015.csv', encoding="windows-1252")
xincomes = list(income['Median Income'].dropna())
yincomes = []
for xincome in xincomes:
    try:
        yincomes.append(int(xincome))
    except ValueError:
        pass
sns.distplot(yincomes, bins=30)
plt.title('median income distribution of US Cities')
plt.xlabel('Median Income of Cities($)')
plt.ylabel('Density')
plt.show()
```

```python
sns.violinplot(yincomes, bins=30)
plt.title('median income Violin Plot of US Cities')
plt.xlabel('Median Income of Cities($)')
plt.ylabel('Density')
plt.show()
```
