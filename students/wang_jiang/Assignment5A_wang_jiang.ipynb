##### Descriptive analysis for the dataset

For this question, I downloaded the dataset called "Tobacco Use 1995-2010 (Prevalence and Trends: Four Level Smoking Data)". (url: https://www.kaggle.com/cdc/tobacco-use/data). I am interested in the percentage of people who somke everyday from 1995 to 2010.

```python
import numpy as np
import pandas as pd
from ggplot import *
```

```python
df = pd.read_csv('tobacco.csv')
df = df.loc[df['State'] == 'Illinois']
```

```python
ggplot(aes(x='Year', y='Smoke everyday'), data=df) +\
    geom_line(color = "blue", size = 1) +\
    geom_point(size = 10) +\
    labs(title = f'Percentage of People Who Smoke Everyday in IL (1995 - 2010)',
       x = r'Year $t$', y = r'Percentage (%)')
```

This plot shows that the percentage of people who smoke everyday in Illinois has decreased sharply since 1996. The decrease is very significant from 2002 to 2010, during which period over 6% of the total population in Illinois reduce their smoking behavior (refrain from smoking everyday). This graph shows that the anti-smoke campaign has made significant progress in IL over the past twenty years.
