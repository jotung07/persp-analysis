```python
x = np.linspace(0, 10, 50)

sin, = plt.plot(x, np.sin(x), '-.xc', label = "Sine") # -. is the line style, x is shape, and c is color
cos, = plt.plot(x, np.cos(x), '--dr', label = "Cosine") # -- is the line style, d is shape, and r is color
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Trig Functions')
plt.legend(handles=[sin, cos])

plt.show()
```

```python
plt.scatter(xScatter, yScatter)
plt.title('Independent Standard Normals')
plt.show()
print("alpha value is:", alpha)
print("beta value is:", beta)
```
