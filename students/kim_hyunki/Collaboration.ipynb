```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('free_throws.csv')
dataset['score_1'], dataset['score_2'] = dataset['score'].str.split('-', 1).str
dataset['score_1'] = dataset['score_1'].convert_objects(convert_numeric=True)
dataset['score_2'] = dataset['score_2'].convert_objects(convert_numeric=True)
```

```python
dataset
```

```python
# Game is tight when the score difference after the free throw is less than 2
is_tight = (-2 < dataset['score_1'] - dataset['score_2']) & (dataset['score_1'] - dataset['score_2'] < 2)

# Game is near end when the remaining time is less than 10 seconds in 4th quarter or in overtime.
is_near_end = ((dataset['period'] >= 4.0) & (dataset['time'].str.startswith("0:0")))
```

```python
No_Pressure = dataset[~is_tight]
No_Pressure.shape
```

```python
Pressure = dataset[is_tight & is_near_end]
Pressure.shape
```

```python
def free_throw_success_rate(subset):
    counts = subset["shot_made"].value_counts()
    return counts[1] / (counts[1] + counts[0])
```

```python
print("Success rate without pressure: " + "%.2f" %free_throw_success_rate(No_Pressure))
print("Success rate with pressure: " + "%.2f" %free_throw_success_rate(Pressure))
```

```python
objects = ('Without pressure', 'With pressure')
y_pos = np.arange(len(objects))
performance = [free_throw_success_rate(No_Pressure), free_throw_success_rate(Pressure)]

%matplotlib inline
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylim(0, 1)
plt.ylabel('Success rate')
plt.title('Effect of stress on free throws')
```

    B. R. Pope and N. G. Pope, “Own-nationality bias: Evidence from UEFA Champions League football referees,” Economic Inquiry, vol. 53, no. 2, pp. 1292– 1304, 2015.
    
    This paper found that referees tend to give favorable calls to the players from their home country in the UEFA Champions League. The measurement used in this paper is net beneficiary calls (fouls committed minus foul suffered) for each player. I think the measurement can be improved if researchers can use human computation power. For example, researchers can give a button to a soccer fans watching the match, which can be pressed whenever they think the foul is unfair. Then, researchers can analyze how many controversial calls each players commits or suffers instead of analyzing every foul calls.
    
    Analyzing all foul calls could be biased since the net beneficiary calls might be correlated with the player's position which again can be correlated to the player's nationality. Attackers are likely to suffer from a foul rather than committing a foul. If a country happens to have proprotionally more attackers than other positions in the UEFA Champions League, referee from this country would seem to give favorable calls to players from its nation even when he is actually being fair. However, if the measurement were only for the controversial fouls, it can minimize this type of systemetic bias.
