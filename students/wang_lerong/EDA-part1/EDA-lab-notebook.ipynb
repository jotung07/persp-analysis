```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline 
df = pd.read_csv("gss2012.csv")
```

```python
df
```

### Univariate analysis
we first take a look at the distribution and density estimation of some single variable in this dataset to understand the composition of respondents of this survey.

```python
df['age'].hist()
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('age distribution histogram')
plt.show()
```

```python
df['sibs'].hist()
plt.xlabel('sibs')
plt.ylabel('frequency')
plt.title('sibs distribution histogram')
plt.show()
```

```python
df['marital'].hist()
plt.xlabel('marital')
plt.ylabel('frequency')
plt.title('marital distribution histogram')
plt.show()
```

```python
df2 = pd.DataFrame({'female': df['age'][df['sex']=="Female"], 'male': df['age'][df['sex']=="Male"]}, 
                   columns=['female', 'male'])
df2.plot.hist(alpha=0.5)
plt.xlabel('age')
```

As we can see from this histogram, there are more female respondents in this dataset. Respondents' age has range from 10s to 80s, and most respondents are married.

```python
# density plot of tv hours
df["tvhours"].plot.density()
```

```python
# happiness according to different ages and genders
sns.swarmplot(x="happy", y="age", hue="sex", data=df)
```

```python
# happiness distribution of divorced people
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y=df["happy"][df["marital"] == "Divorced"], data=df, color="c")
```

Most people seem to be pretty happy, but less are very happy

```python
# happiness distribution of married people
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y=df["happy"][df["marital"] == "Married"], data=df, color="c")
```

Most people are pretty happy, and a larger proportion of people, compared to divorced people, are very happy. 

```python
# happiness distribution of divorced female
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y=df["happy"][(df["marital"]=="Divorced") & (df["sex"]=="Female")], 
              data=df, color="c")
```

```python
# happiness distribution of divorced male
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y=df["happy"][(df["marital"]=="Divorced") & (df["sex"]=="Male")], 
              data=df, color="c")
```

There's no apparent pattern shows that there is a difference between male and female regarding to their happiness status after divorce

```python
# bar chart of male and female divorced individuals' happiness distribution
sns.countplot(x=df["happy"][df["marital"]== "Divorced"], hue="sex", data=df, palette="Greens_d")
```

```python
# bar chart of male and female married individuals' happiness distribution
sns.countplot(x=df["happy"][df["marital"] == "Married"], hue="sex", data=df, palette="Greens_d")
```

In comparing the last two bar charts, I think generally speaking, married individuals are happier than divorced individuals

### Some other explorations

```python
sns.factorplot(x="age", y="gunlaw", hue="sex",
               col="race", data=df, kind="swarm")
```

Most people, no matter what races they are, favor the gun permits.

```python
sns.factorplot(x="age", y="polviews", hue="sex",
               col="race", data=df, kind="swarm")
```

In all three race categories, more people consider themselves as having moderate political views, and most black people and people of other races perceive themselves as moderate.
