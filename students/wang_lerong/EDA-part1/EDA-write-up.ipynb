## Does happiness correlate with sex and race?

From the plots below, we can see that most people seem to be "pretty happy" with their current life. I first generated bar plots to compare the count of each of the three happiness status, which are shown in the lab notebook. However, since the number of respondents of different races and sex differs in this dataset, so with the bar plots showing the count in each category, it's hard for us to directly do some comparisons, so I then generated pie charts. For sex, there’s no apparent differences between the proportion of three happiness categories for both male and female. I may conclude that happiness does not correlate with sex. For race, I mainly compared happiness distribution between white and black, since there are only three categories (white, black and others) in this dataset. I found white people and black people appear to have similar proportions for "pretty happy". However, it seems that the proportion of "very happy" among white people is larger than that of black people, and the proportion of "not too happy" among black people is larger than that of white people. The overall proportion of pretty happy and very happy for white is 87.3%, and the overall proportion of pretty happy and very happy for black is 79.6%, so from this dataset, we may conclude that white people are happier than black people.

```python
df["happy"][df["sex"]=="Female"].value_counts()
```

```python
labels = 'Not too happy', 'Pretty happy', 'Very happy'
sizes = [154, 592, 336]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie chart of happiness among female')
plt.show()
```

## Does happiness correlate with marital status? Are there any differences between men and women?

I am especially interested in how marital status affect people’s happiness and whether there’s any differences between how men and women react in different marital status. The marital status in the original dataset has five categories, which include never married, married, separated, widowed, and divorced. In this section, we only make a comparison between married and divorced. More people under both conditions appear to be pretty happy about their current life. Also, we can see from the bar plot that more married individuals are “very happy” than divorced individuals, less married individuals are “not too happy” than divorced individuals, and most individuals are “pretty happy” no matter they are married or divorced, so generally speaking, married individuals are happier than divorced individuals. In this sense, we may conclude that happiness does correlate with marital status. Then we want to know whether there’s a difference between male and female in how marital status affects happiness. The proportion of female in each happiness category under married condition and divorced condition is similar to that of male, so gender does not play an important role in determining happiness under different marital status. 

```python
sns.countplot(x=df["happy"][df["marital"]== "Divorced"], hue="sex", data=df, palette="Greens_d")
plt.title('bar plot of frequency of each happiness status among divorced individuals')
```

```python
sns.countplot(x=df["happy"][df["marital"] == "Married"], hue="sex", data=df, palette="Greens_d")
plt.title('bar plot of frequency of each happiness status among married individuals')
```

## Does happiness correlates with social connect and social trust? Are there any differences between men and women

Finally, I explored the relationship between happiness and social trust and happiness and social connect, because it seems to me that people with more social trust and social connect are probably more optimistic and outgoing, so they may appear to be happier. I can make a hypothesis that happiness correlates positively with both social connect and social trust. While exploring this problem, I generated two bar plots. For social connect, we can see from the bar plot that for both male and female, individuals who are happier also tend to have higher social connect. The patterns look similar for both male and female, so we can say that social connect has a positive relationship with happiness, and there’s no apparent difference between male and female. For social trust, it appears that for female, higher social trust indicates a happier status. For male, though the social trust levels of very happy and pretty happy are very close, and probably if the number of observations in the dataset grows, there will be certain pattern that we can conclude, we cannot conclude from this dataset that happiniess correlates positively with social trust for male.

```python
sns.barplot(x="sex", y="social_connect", hue="happy", data=df)
plt.title('bar plot of the relationship between social connect and happy')
```

```python
sns.barplot(x="sex", y="social_trust", hue="happy", data=df)
plt.title('bar plot of the relationship between social trust and happy')
```
