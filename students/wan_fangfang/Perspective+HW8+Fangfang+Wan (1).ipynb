```python
##Perspective HW8 Fangfang Wan
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

college = pd.read_csv('College.csv')
USArrests = pd.read_csv('USArrests.csv')
college.drop(['Private'], axis=1, inplace=True)
```

```python
### Colleges Problem
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
college_pca = pca.fit_transform(college)
```

```python
college_pca
```

```python
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
```

```python
plt.scatter(college_pca[:, 0], college_pca[:, 1], color='red')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of college dataset')
```

```python
print('\nEigenvalues \n%s' %pca.explained_variance_)
print('Eigenvectors \n%s' %pca.components_)
```

```python
max1 = max(pca.components_[0])
lst1 = []
for i in pca.components_[0]:
    if i != max1:
        lst1.append(i)
print(max(lst1))
```

```python
max2 = max(pca.components_[1])
lst2 = []
for i in pca.components_[1]:
    if i != max2:
        lst2.append(i)
print(max(lst2))
```

```python
#Clustering states problem: Question 6: 
# needed imports
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
%matplotlib inline
np.set_printoptions(precision=5, suppress=True)
```

```python
Z = linkage(USArrests, 'ward')
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram for USArrests Dataset')
plt.ylabel('distance')
dendrogram(Z, leaf_rotation=90., leaf_font_size=12)
plt.show()
```

```python
def fancy_dendrogram(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)

    ddata = dendrogram(*args, **kwargs)

    if not kwargs.get('no_plot', False):
        plt.title('Hierarchical Clustering Dendrogram (truncated)')
        plt.ylabel('distance')
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x, y, 'o', c=c)
                plt.annotate("%.3g" % y, (x, y), xytext=(0, -5),
                             textcoords='offset points',
                             va='top', ha='center')
        if max_d:
            plt.axhline(y=max_d, c='k')
    return ddata
fancy_dendrogram(
    Z,
    truncate_mode='lastp',
    p=12,
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,
    annotate_above=150,
    max_d=200 # plot a horizontal cut-off line
)
plt.show()
```
