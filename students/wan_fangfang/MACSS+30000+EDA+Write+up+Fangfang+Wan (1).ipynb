1. Degree and income
    From the scatter plot, we can see that for people in every degree group, there is a significant amount of people who earn very high income, which means that a low degree does not prevent people from earning high income. The density of earning hign income does not differ much according to degree level, at least under the precision of the scatter plot. However, for people with higher degrees, it is quite unlikely for them to earn very low income. Their lower bound is much higher than that of people with lower degrees. This may result from the fact that a higher degree is always a signal of higher ability, and employees tend to hire people with higher degrees. There are some outliers, however, in which people with graduate degree earn very low, but we're not clear about the reasons behind it. There could be many reasons, such as health shock, or the person was pursuing a Ph.D. degree in 2006, so his or her income was low in 2006, but his expected income in the future should be high. For people with all degrees, it is always hard for them to earn extremely high income. Those people with high income can also be considered as outliers. However, the possibility of earning extremely high income is not very different for people with lower degrees and higher degrees. This may result from the fact that some people with low degrees have excellent working skills, but they are not good at studying materials at school. They might also have talents in their specialties, which can bring high income to them. 
    

2. Age and income
   From the scatter plot, we find simlar patterns as in the relationship between degree and income - it is far less likely for older people to earn extremely low income, because as people get older, experience in their expertise tend to bring positive returns as income. Therefore, it is very unlikely for older people to stay in the basic working positions after years. The lower income bound for older people tend to be much higher than that of younger people. There are some outliers, however, in which older people earn very few income, but that situation occurs when the person is over 80, when any person can lose a significant amount of intellectuality. However, it is also far more likely for younger people to earn very high income, although it is hard for people at all ages to earn extremely high income. Such situation might be resulted from the fact that it is more likely for young people to take risks, which can bring higher standard deviation but also higher upper bound. In contrast, older people tend to seek stability due to their various responsibilities, such as family, children, etc. Therefore, the possible income range for young people is much wider than that of old people, and statistically, it means that the standard deviation of income for young people is higher than that of older people. 
   
3. About missing values
    There are some missing values for income06, and from observing their characteristics in the dataset, it shows the pattern that those are mostly people who do not have a full-time job, which can be people who do not have a formal job, such as students, who are keeping house, who are unemployed, etc.
