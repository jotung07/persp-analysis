### Age and libertarianism

My EDA has sparked my interest in how on government intervention in contentious social issues vary across age groups. There is an assumption that opinion on social issues is becoming 'more progressive' over time. For example, there was a decent amount of enthusiasm for Gary Johnson's Libertarian's during the 2016 Presidential Campaign from young people, as well as of course for the Democrats. However, the data on social issues seems to present a mixed picture.

The two variables I focus on are 'gunlaw' (FAVOR OR OPPOSE GUN PERMITS) and 'pornlaw'/'pornlaw2' (FEELINGS ABOUT PORNOGRAPHY LAWS - essentially should it be legal. 'pornlaw2' combines legal for over 18 and legal for all). In the strict negative sense of the word, the 'liberal' position on both fronts is to oppose government intervention (i.e. 'gunlaw' == oppose and 'pornlaw2' == Not illegal to all). However, 'liberal' in the political sense of Democrat policy is to lean towards favoring gun permits. Also, there may be a feminist counterargument to porn legality, which may lead some millenials to actually favor porn being illegal.


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("data/gss2012.csv")
```

Let's compare swarmplots for how age affects opinion on the two variables.

```python
sns.swarmplot(x="pornlaw", y="age", hue = 'race', data=df).set_title('Swarm plot of porn and age, with race')
```

```python
sns.swarmplot(x="gunlaw", y="age", hue = 'race', data=df).set_title('Swarm plot of guns and age, with race')
```

These graphs show that there is some pattern: the youth do seem to favor the progressive opinion. 
A research question one could investigate is the'Naive Millenial Hypothesis': that under 40s favor 'more progressive' opinions on social questions. You could also augment this by asking how this changes depending on social question. For example, it seems that young people are much less likely to want to ban porn than protect the Second Amendment. One could also ask whether other variables interact with these variables. It seems from this graph that race is not a significant variable for example.

However, the graph also suggests a more mixed picture with a resurgence in Second Amendment support amongst Millenials in their twenties. It would be interesting to investigate which covariates are causing this resurgence.

```python
sns.swarmplot(x="pornlaw2", y="age", hue = 'relig', data=df).set_title('Swarm plot of porn and age, with religion')
```

From this graph it seems that there is a cluster of beige dots towards the bottom of 'not illegal to all', which suggests that correlation to 'None' religion could explain the youth's more 'liberal' opinion on porn.

```python
df['gunlaw'].value_counts(dropna = False).plot(kind = 'bar')
plt.xlabel('Response')
plt.ylabel('Frequency')
plt.title('Missingness in gun permit opinion')
```

As can be seen 'gunlaw' is one of the categories which I discovered were optional questions because they have a significant amount of missingness. Therefore, in our analysis of these social issues, we should be wary of conclusions that we draw if missingness is systematically biased with an underlying opinion. For example, it is possible that those who choose not to answer this government survey might correlate with libertarian opinion so we may be underestimating the opposal of gun permits.

If an analyst chooses to look at how TV watching habits interact with these social issues they should exclude extreme outliers (e.g. 24 hours TV), as they are probably not answering truthfully in general and are adding noise to the analysis.
