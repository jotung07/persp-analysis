 ### 3. Descriptive plot that highlights some characteristic of the data
 ### Pokemon with stats (https://www.kaggle.com/abcsds/pokemon)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv("/Users/ivychen/persp-analysis/students/chen_ruxin/Pokemon.csv")
```

```python
pokemon.head()
```

```python
%matplotlib inline
plot = sns.factorplot(x="Type 1", y="Total", hue="Legendary", col= "Generation", data=pokemon, kind="bar", col_wrap=2, size=10)
plt.subplots_adjust(top= 0.95)
plot.fig.suptitle(r'Barplot of General Statistics for Pokemon', fontsize = 20)
```

It can be seen from the second graph that the legendary Pokemon have higher total scores than their non-legendary counterparts. The total scores are between 300 and 500 for most of the types across different generations. The total score of generation 4 seems to be the highest, and the variances between types are larger for generation 2 for non-legendary Pokemon.
