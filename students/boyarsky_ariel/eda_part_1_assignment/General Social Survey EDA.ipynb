### How People Get Ahead and Income Level

Looking to our exploratory analysis of the General Social Survey data above, we find multiple phenomena worth a further examination. We began our analysis with simple graphs depicting the variation with variables that caught our attention. These graphs on their own do not represent the most striking features of our analysis. However, we use these as entry points for further analysis on the covariation of groups of variables. This reveals a few interesting trends that spur us to dig deeper. Specifically, we are interested in how respondents answered the “getting ahead question.” This question asks people how they think others get ahead. Respondents are given the choice of Hard Work, Luck or Help, or Both Equally. First, we analyze the variation in some key variables, then we conduct sub group analysis to understand how various groups respond to this question. The key finding, we see is that people generally respond that hard work is the way people get ahead regardless of income, education, and political views. While this effect increases with some group, every group in the sample favors Hard Work over Luck or Help.

```python
dta['getahead'].value_counts().plot(kind='bar')
plt.ylabel("Count")
plt.xlabel("Responses")
plt.title("Variation in Response to 'How People Get Ahead?'")
plt.show()
```

This first graph shows the variation in the getting ahead question. As we see from above there is a clear bias towards Hard Work in our sample. Indeed, Luck or Help receives the least responses at around 100.

The next two plots show similar distributions for the income levels question, political views, and education. We use bar plots to display these distributions. Notice, that income levels are based in response to the real income question. This variable conveniently puts these responses in categories from low to high. This is helpful for our analysis.

```python
dta['rincom06_5'].value_counts().plot(kind='bar')
plt.ylabel("Count")
plt.xlabel("Income Level")
plt.title("Variation in Income Level")
plt.show()
```

```python
dta['educ'].value_counts().plot(kind='bar')
plt.ylabel("Count")
plt.xlabel("Education Level")
plt.title("Variation in Sample Education Level")
plt.show()
```

Note that in our final graph of education level we see that the distribution is skewed towards respondents who finished high school, and had some college. This is reflective of the general populace of the United States. We do have some outliers such as those who only completed 1st and 2nd grade.

Now that we have examined the variation in each variable we look at the how these variables vary when grouped together. Since we have categorical variables we turn to the package Seaborn which has a few graphs that help us visualize the variation between categorical variables. Specifically, we use the swarmplot which allows us to plot two categorical variables and a third quantitative variable. We use this final variable to plot age, which may be relevant to political views, education, and income.

Our first swarmplot visualizes answers to getting ahead according to income level and age. Notice, that green dots (that represent Hard Work) are very prevalent in every level of income. Interestingly, we see the largest number of blue dots (representing Both Equally, that is both luck and hard work) in the highest level of income. This is also very interesting as this is counterintuitive to classical wisdom. Why many high-income individuals emphasize both luck and hard work may be an interesting question. Also notice most respondents are between 25 and 65 years old. There is no clear correlation by age in either variable.


```python
s.swarmplot(x="age", y="rincom06_5", hue="getahead", data=dta)
plt.xlabel("Age")
plt.ylabel("Income Group")
plt.title("Covariation of Income Level, Getting Ahead Response, and Age")
plt.legend(title="How Poeple Get Ahead Response", bbox_to_anchor=(1, 1))
plt.show()
```

This next swarmplot shows the distribution of responses to how people get ahead based on age and political views. Again, we see that political views seem to have no effect on people’s response to getting ahead. Indeed, we again see many green dots in each category. As per our conventional expectations, we see the most variation in the extreme liberal group. However, even here most dots are green. Again, there is no clear correlation by age.

```python
s.swarmplot(x="age", y="polviews", hue="getahead", data=dta)
plt.xlabel("Age")
plt.ylabel("Political Views")
plt.title("Covariation of Political Views, Getting Ahead Response, and Age")
plt.legend(title="How Poeple Get Ahead Response", bbox_to_anchor=(1, 1))
plt.show()
```

The next plot we show to examine how respondents answer the getting ahead question uses a grouped bar plot to examine two categorical variables. This plot looks at the distribution of responses to getting ahead based on education level. As expected those with 4 years of college are the group that chooses Hard Work the most. After this, we see those that completed high school, and had some college. This graph shows more of a difference between those with higher and lower education levels. However, since there are very few observations with less than a high school education it is difficult to make systematic inferences about the preferences of this group.

```python
s.countplot("getahead", hue="educ", data=dta)
plt.xlabel("How People Get Ahead Response")
plt.ylabel("Count")
plt.title("Distribution of Getting Ahead Response by Education Level")
plt.legend(title="Education Level", bbox_to_anchor=(1, 1))
plt.show()
```

Now that we have completed our analysis of variation between variables we conduct an analysis of outliers (we omit an analysis of missing data since our EDA showed no crucial findings). Below is a bar plot which was built on a frequency table of income level and getting ahead response. We focus on these two variables since we have a good sample size for each category. However, when we look at a frequency table of the two variables we find that we have very few observations with a Luck or Help response in the low, middle, and high-income groups. We point this out because these outliers may influence our results in future analysis. Thus, we should consider this carefully in future analyses. Note, that the 15 count threshold was arbitrary. A future analysis should use a more sophisticated technique such as Cook’s distance to confirm possible high influence points. Also, this underscores that we have very few individuals who responded Luck or Help to the getting ahead question, even in the low-income group. 

```python
freq.size()[freq.size() < 15].plot(kind="bar")
plt.xlabel("Income Level and Getting Ahead Response")
plt.ylabel("Count")
plt.title("Possible Outliers in the Variation of Getting Ahead Response and Income Level with Counts below 15")
plt.show()
```

Finally, we conclude our analysis with a correlation matrix for income level and getting ahead as well as political views and getting ahead. We do this by creating dummies of each variable (i.e. 1 or 0 columns for each category). Interestingly, we do not see any strong correlations in income level and response. This is probably because we have divided our data in subgroups for analysis. However, this is an interesting starting point for a deeper analysis of how people respond to the getting ahead question. To conclude, we have explored how participants of the general social survey have responded in an atypical fashion to the question of how other people get ahead. Conventional wisdom may assert that those with higher incomes or education levels would value hard work while respondents on the lower level of the spectrum may consider the combination of hard work and luck or just luck and help. However, we see the opposite trend with most people who stress both hard work and luck being members of the highest income bracket. Additionally, we see respondents answer hard work across the spectrum. This opens an interesting sociological examination of how people view success and what causes success. Indeed, this may play into beliefs about the ‘American Dream’ or other cultural phenomena. Regardless of cause, this anomaly represents an interesting question in socio-economic research that is worth closer examination.
