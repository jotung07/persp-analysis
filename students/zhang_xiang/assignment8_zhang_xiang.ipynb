```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
# Read in data
collegedata = pd.read_csv("College.csv")

# Select numeric variable
collegedata = collegedata.select_dtypes(['number'])
```

```python
# Change data to DataFrame
collegedata = pd.DataFrame(collegedata, columns = collegedata.columns, index = None)
```

```python
# Do PCA
pca = PCA()
pca.fit(collegedata)
```

```python
# Plot the factor loadings of each variable
column_names = ["PC" + str(x + 1) for x in range(0, pca.n_components_)]
pca_factorloading = pd.DataFrame(pca.components_.T, index = collegedata.columns, columns = column_names)
pca_factorloading
```

Now we plot the first two components

```python
pca_factor_loading_2 = pd.DataFrame(pca.components_.T[:, :2], index = collegedata.columns, columns = ['PC1', 'PC2'])
```

```python
reduced_data = pd.DataFrame(pca.transform(collegedata), columns = column_names)
```

```python
def vector_plot(data, pca_loading, num_factor, xmin, xmax, ymin, ymax):
    
    fig, ax = plt.subplots(figsize = (14,8))

    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    
    feature_vectors = pca.components_.T

    # using scaling factors to make the arrows
    arrow_size, text_pos = 7.0, 8.0,

    for i in range(num_factor):
        ax.arrow(0,0,pca_loading.PC1[i], pca_loading.PC2[i])
    
    for i in pca_loading[['PC1', 'PC2']].index:
        ax.annotate(i, (pca_loading.PC1.loc[i] * 1.05, pca_loading.PC2.loc[i] * 1.05), color='red')
    
    ax.set_xlabel("Principal Component 1", fontsize=14)
    ax.set_ylabel("Principal Component 2", fontsize=14)
    ax.set_title("Principal Coponent Plane with First Two Components", fontsize=16);
    return ax

vector_plot(collegedata, pca_factor_loading_2, 17, -0.1, 1, -0.4, 1)
plt.show()
```

From the figure above, combined with the factor loading table, we get that:
- The first component is most correlated with `F.Undergrad`, `Apps`, `Accept`, `Enroll` (just select four variables most relavant).  
- The second component is most correlated with `Expend`, `Outstate`, `Room.Board`, `F.Undergrad`. (just select four variables most relavant).  

I also plot the observations on their first two components.

```python
fig, ax = plt.subplots(figsize = (14,8))
    
ax.scatter(x = reduced_data.loc[:, 'PC1'], y=reduced_data.loc[:, 'PC2'], facecolors='b', edgecolors='b', s=70, alpha=0.5)
ax.set_xlabel("Principal Component 1", fontsize=14)
ax.set_ylabel("Principal Component 2", fontsize=14)
ax.set_title("Principal Component plane with original feature projections.", fontsize=16)

plt.show()
```

### Colleges Problem 2

```python
var_exp = pca.explained_variance_ratio_ * 100
cum_var_exp = np.cumsum(var_exp)
x = range(1, pca.n_components_ + 1)
```

I first plot variance explained by each component separately

```python
# Variance explained by each component
plt.plot(x, var_exp, '-o', label = 'Single Factor ')
plt.title('Variance Explained by Each Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Percentage (%)')
plt.xlim(0, pca.n_components_ + 1)
plt.show()
```

I then draw the cumulative variance explained.

```python
# Comulated variance explained
plt.plot(x, cum_var_exp, '-o', label = 'Single Factor ')
plt.title('Cumulative Variance Explained')
plt.xlabel('Principal Component')
plt.ylabel('Percentage (%)')
plt.xlim(0, pca.n_components_ + 1)
plt.show()
```

```python
cum_var_exp
```

From the figure and the plotted list, We can see that **87.08%** variance is explained by the first two principal components.

### Cluster States Problem 2 - Problem 5

I Perform $K$-means clustering with $K=2$, $K=3$, $K=4$ using raw data, and $K$-means clustering with $K=3$ using the first two principal components score vectors below.

```python
from sklearn.cluster import KMeans
```

Below I first write down the function for plotting the clustering and plot the four figures.

```python
def kmeans_cluster(data, n, ax, title):
    km = KMeans(n_clusters = n).fit(data)
    cluster = km.labels_
    centers = km.cluster_centers_
    centers = pca.fit_transform(centers)
    ax.scatter(x = reduced_data["PC1"], y = reduced_data["PC2"], marker = '.', c = cluster)
    ax.scatter(x = centers[:, 0], y = centers[:, 1], marker = "x", s = 150)
    ax.set_title(title)
    ax.set_xlabel("First Principal Component")
    ax.set_ylabel("Second Principal Component")
    
figure, axes = plt.subplots(2, 2, figsize = (12,12))

kmeans_cluster(arrestdata, 2, axes[0, 0], "K = 2, Raw Data")
kmeans_cluster(arrestdata, 3, axes[0, 1], "K = 3, Raw Data")
kmeans_cluster(arrestdata, 4, axes[1, 0], "K = 4, Raw Data")
kmeans_cluster(reduced_data, 3, axes[1, 1], "K = 3,  First two principal components score vectors")

plt.show()
```

### Cluster States Problem 6

Below I cluster the states using hierarchical clustering with complete linkage and Euclidean distance

```python
import scipy.cluster.hierarchy as hr
```

```python
hie_cluster = hr.linkage(arrestdata, method='complete', metric='euclidean')

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
dn = hr.dendrogram(hie_cluster, labels = arrestdata.index, leaf_font_size = 10)
plt.title('Hierarchical Clustering of States: Complete Linkage and Euclidean Distance')
plt.show()
```

### Cluster States Problem 7

Below I cut the dendrogram at a height and group states in three distinct clusters

```python
hie_cluster = hr.complete(arrestdata)

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
dn = hr.dendrogram(hie_cluster, truncate_mode='lastp', p = 3, labels = arrestdata.index)
plt.title('Hierarchical Clustering of States: Complete Linkage and Euclidean Distance')

plt.show()
```

All states have been grouped into three groups:
- First group (16 states): Alabama, Alaska, Arizona, California, Delaware, Florida, Illinois, Louisiana, Maryland, Michigan, Mississippi, Nevada, New Mexico, New York, North Carolina, and South Carolina.
- Second group (14 states): Arkansas, Colorado, Georgia, Massachusetts, Missouri, New Jersey, Oklahoma, Oregon, Rhode Island, Tennessee, Texas, Virginia, Washington, and Wyoming.
- Third group (20 states): Connecticut, Hawaii, Idaho, Indiana, Iowa, Kansas, Kentucky, Maine, Minnesota, Montana, Nebraska, New Hampshire, North Dakota, Ohio, Pennsylvania, South Dakota, Utah, Vermont, West Virginia, and Wisconsin.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler(with_std = True)
```

```python
arrestdata_scaled = pd.DataFrame(scaler.fit_transform(arrestdata), index = arrestdata.index, columns = arrestdata.columns)
```

```python
hie_cluster = hr.linkage(arrestdata_scaled, method='complete', metric='euclidean')

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
dn = hr.dendrogram(hie_cluster, labels = arrestdata_scaled.index, leaf_font_size = 10)
plt.title('Hierarchical Clustering of States: Complete Linkage and Euclidean Distance')
plt.show()
```

Comparing the graphs with and without scaling, we can see that:
- After normalizing, the distance has been reduced. 
- After normalizing, we're making the variables "comparable". Conceptually, we should scale variables before clustering because different variables have different unit and different variance, so they're not directly comparable. If we didn't scale the variables, the variance of certain "big variance" variables may lead to a undesirable clustering results.
