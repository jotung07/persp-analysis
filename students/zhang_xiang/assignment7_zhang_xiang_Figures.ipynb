```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gss = pd.read_csv("gss2012.csv")
```

### Basic Distribution of Variables

```python
# Age Distribution
bins = np.arange(25, 81, 5)
agedata = gss["age"][gss["age"] <= 81]
plt.hist(agedata, bins = bins, alpha=0.5, color='#EDD834', ec='black')
plt.title('Age Distribution')
plt.show()
```

```python
# Generate Box Graph of Age
gss.boxplot("age")
plt.show()
```

```python
# Generate age group
agecut = [25, 40, 60, 80]
gss["agegroup"] = pd.cut(gss["intage"],\
						 bins = agecut,\
						 labels = ["25 - 40", "41 - 60", "61 - 79"],\
						 include_lowest = True, right = False)
```

```python
# AGe group Distribution
vc = gss["agegroup"].value_counts()
vc.plot(kind='bar', subplots=True, color = '#EDD834')
plt.title('Age Group')
plt.show()
```

```python
# Number of siblings
bins = np.arange(0, 31, 3)
sibsdata = gss["sibs"][gss["sibs"] <= 30]
plt.hist(sibsdata, alpha=0.5, color='#EDD834', ec='black', bins = bins)
plt.title('Number of Siblings Distribution')
plt.show()
```

```python
# Generate Box Graph of NUmber of Siblings
gss.boxplot("sibs")
plt.show()
```

```python
# Generate Box Graph of NUmber of Siblings, ignoring outliers
gss.boxplot("sibs", showfliers=False)
plt.show()
```

```python
# Number of children
bins = np.arange(0, 9, 1)
childsdata = gss["childs"][gss["childs"] <= 30]
plt.hist(childsdata, alpha=0.5, color='#EDD834', ec='black', bins = bins)
plt.title('Number of children Distribution')
plt.show()
```

```python
# Generate Box Graph of NUmber of childrens
gss.boxplot("childs")
plt.show()
```

```python
# Generate Box Graph of NUmber of childrens, ignoring outliers
gss.boxplot("childs", showfliers = False)
plt.show()
```

```python
# Gender distribution
vc = gss["sex"].value_counts()
vc.plot(kind='bar', subplots=True, color = '#EDD834')
plt.show()
```

```python
# Gender distribution, change to pie chart
plt.pie(vc, labels = ["Female", "Male"], shadow=False, colors = ["#E13F29", "#D69A80"], startangle=90,)
plt.show()
```

```python
# Distribution of Income Group
vc = gss["income06"].value_counts()
vc.plot(kind='bar', subplots=True, color = '#EDD834')
plt.title("Histogram of Household Income")
plt.show()
```

```python
# Distribution of Income Group (rough one)
vc = gss["rincom06_5"].value_counts()
vc.plot(kind='bar', subplots=True, color = '#EDD834')
plt.title("Histogram of Household Income")
plt.show()
```
