### Background and Introduction
As noted in the General Social Survey questionnaire, "We are faced with many problems in this country, none of which can be solved easily or inexpensively". To better allocate resources to places where social welfare can be improved, it's important to know about the general public's attitude towards the amount of money spent in certain areas.

In this assignment, I try to explore people's attitudes toward the budget on four social problems: "Foreign Aid", "Military, Armaments, And Defense", "Assistance for Childcare", and "Solving Problems of Big Cities". 

### Heterogeneous Attitudes toward the Budget for Different Fields
To show the heterogeneity in people’s attitudes toward the budget on different issues, I draw four bar graphs depicting the distribution of attitudes.

We can see from these four figures that people have quite different attitudes on the budget. We can get two facts here. First, people’s choices are quite different, since an “economic significant” proportion of people choose “Too much”, “Too little”, and “About right” (which means that the proportion itself is economically significant). Besides, the general attitudes toward the budget for the four issues are quite different. As we can see, the share of people thinking that government spend too much on foreign aid has the largest share, while most people agree that they're spending the right amount on military, and most people think they just spend too little on "assistance for children" and "solve problems of big cities". From these four graphs, I show that people have very different attitudes toward the money spent in different fields.


```python
# Import data and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

gss = pd.read_csv("gss2012.csv")
```

```python
# Distribution of "nataid"
plt.subplot(2, 2, 1)
vc = gss["nataid"].value_counts()
vc.plot(kind='bar', subplots=True)
plt.title("Foreign Aid")
plt.ylabel("Count")
plt.xlabel("Attitude")

# Distribution of "natarms"
plt.subplot(2, 2, 2)
vc = gss["natarms"].value_counts()
vc.plot(kind='bar', subplots=True)
plt.ylabel("Count")
plt.title("Military, Armaments, And Defens")
plt.xlabel("Attitude")

# Distribution of "natchld"
plt.subplot(2, 2, 3)
vc = gss["natchld"].value_counts()
vc.plot(kind='bar', subplots=True)
plt.ylabel("Count")
plt.title("Assistance For Childcare")
plt.xlabel("Attitude")

# Distribution of "natcity"
plt.subplot(2, 2, 4)
vc = gss["natcity"].value_counts()
vc.plot(kind='bar', subplots=True)
plt.ylabel("Count")
plt.title("Solving Problems Of Big Cities")
plt.xlabel("Attitude")

plt.tight_layout()
plt.suptitle('Attitudes toward Amount of Money Spent on Different Fields')
plt.subplots_adjust(wspace=0.8, top=0.9)
plt.show()
```

### Heterogeneous Attitudes by Gender
One natural question that follows is, "Do people with different attitudes toward the budget have different background?". Or more specifically, do certain groups of people tend to have more positive attitudes toward the budget while some other groups hold more negative attitudes? My hypothesis is that people of different gender, age, and socioeconomic status may have different attitudes toward the budget. 

In this section, I explore the heterogeneity from the gender angle. To do so, I draw bar graphs of attitudes toward the amount of money spent on certain fields by gender.

```python
groupvar = "sex"

question = "nataid"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Attitude toward Forien Aid")
plt.show()

question = "natarms"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Military, Armaments, And Defens")
plt.show()

question = "natchld"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Assistance For Childcare")
plt.show()

question = "natcity"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Solving Problems Of Big Cities")
plt.show()
```

From these graphs we can see that the attitudes of female and male are quite different in some fields. For example, about 250 females and about 150 males think that the spending on military is "about the right" (the ratio is 250 / 150 = 1.6). However, about 160 females and about 80 males think the spending on military is "too little" (the ratio is 2 now!), and about the same number of female and male thinks we spend too much on that (the ratio is 1 now!). Thus, we can see that compared with male, female request more military spending. In the same way, we can see that compared with male, female also want more spending on assistance for children and solving problems of big cities.

### Heterogeneous Attitudes by Age Group
Also, I hypothesize that people of different age groups may tend to have different attitudes toward the money spent in certain fields. To test my hypothesis using EDA, I draw similar graphs by age group. In my analysis, I divide people into three groups: 21-40, 41-60, and 61-79.

```python
# Generate age group
agecut = [25, 40, 60, 80]
gss["agegroup"] = pd.cut(gss["intage"],\
						 bins = agecut,\
						 labels = ["25 - 40", "41 - 60", "61 - 79"],\
						 include_lowest = True, right = False)

groupvar = "agegroup"

question = "nataid"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Attitude toward Forien Aid")
plt.show()

question = "natarms"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Military, Armaments, And Defens")
plt.show()

question = "natchld"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Assistance For Childcare")
plt.show()

question = "natcity"
vc = gss.groupby([groupvar, question])[question].value_counts().unstack(groupvar)
vc.plot(kind = 'bar', subplots = False)
plt.xlabel("Attitude")
plt.ylabel("Count")
plt.title("Solving Problems Of Big Cities")
plt.show()
```

In a similar way, we can see that about 250 middle-age individuals think the government spends the right amount on military, and the number of elders (who ages 61-79) is about 150, thus the ratio here is 250 / 150 = 1.6. However, when it comes to the answer "too little", the ratio changes to about 1.3, and the ratio goes back to 1.5 when the answer is "too much". 

From all these figures, we can see that some variation in attitudes do exists, but not as large as the variation between female and male. When we analyze the heterogeneity from the age group perspective, I find that people from different age groups hold most diverging attitudes to the amount of money spent on "solving problems of big cities".

### Heterogeneous Attitudes by Income Level
Socioeconomic status may also affect people's attitudes toward the amount of money spent on certain fields. If people are in different socioeconomic status, they may tend to focus on different issues. 

In this section, I explore the heterogeneity in attitudes between different income levels.

```python
sns.countplot(y="nataid", hue="rincom06_5", data=gss)
plt.ylabel("Attitudes")
plt.title("Attitudes toward the Money spent on Forien Aid")
plt.show()

sns.countplot(y="natarms", hue="rincom06_5", data=gss)
plt.title("Attitudes toward the Money spent on Military")
plt.ylabel("Attitudes")
plt.show()

sns.countplot(y="natchld", hue="rincom06_5", data=gss)
plt.title("Attitudes toward the Money spent on Assistance to Children")
plt.ylabel("Attitudes")
plt.show()

sns.countplot(y="natcity", hue="rincom06_5", data=gss)
plt.title("Attitudes toward the Money spent on Solving big-city problems")
plt.ylabel("Attitudes")
plt.show()
```

From these graphs, we can see that people from different income groups do have different attitudes toward the money spent. From these figures, we can see that the poor people are more likely to say "we spend too little on foreign aid", the middle-class are more likely to be against military spending, and relatively speaking, the poor people want more money to be put in solving big-city problems.
