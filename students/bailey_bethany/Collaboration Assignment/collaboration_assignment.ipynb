### Kaggle Dataset
My dataset consisted of information on post-graduation salaries for college graduates. My graph plots the distribution of the mid-career 10th percentile and 90th percentile salaries for college graduates. I chose this comparison in order to compare the distribution of the 10th percentile and 90th percentile salaries. The data can be found at: https://www.kaggle.com/wsj/college-salaries

```python
import pandas as pd
import matplotlib.pyplot as plt

sal_by_reg = pd.read_csv('salaries-by-region.csv', index_col='Region')
clean_df = sal_by_reg.dropna(axis=0).replace({'\,':''}, regex = True).replace({'\$':''}, regex = True)
ten_num = pd.to_numeric(clean_df['Mid-Career 10th Percentile Salary'], errors='coerce')
```

```python
%matplotlib inline
plt.hist(ten_num, bins=50)
plt.title('Mid-Career 10th Percentile Salary Distribution', fontsize=15)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
ninety_num = pd.to_numeric(clean_df['Mid-Career 90th Percentile Salary'], errors='coerce')
%matplotlib inline
plt.hist(ninety_num, bins=50)
plt.title('Mid-Career 90th Percentile Salary Distribution', fontsize=15)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

As you can see from the graphs above, the distribution of the 90th percentile is similar to the distribution of the 10th percentile. However, the 10th percentile is slightly more distributed, especially at higher incomes.

## Part 2: Improving a Journal Article

Hwang, Jackelyn and Robert J. Sampson. 2014. ["Divergent Pathways of Gentrification: Racial Inequality and the Social Order of Renewal in Chicago Neighborhoods."](http://www.jstor.org/stable/43187561) *American Sociological Review* 79(4): 726-751.

### Paper Summary and Methods

This article systematically used Google Street View (GGO) to identify neighboorhood change and gentrification. The researchers then compared these coded blocks to previously-existing data used to identify gentrification, such as census data, racial breakdown, presence of Starbucks and green roofs, police records, prior street-level observations, measures of amenities, poverty, homeownership, and vacancy, incidents of homicide, City of Chicago data, and a PHDCN survey of percentions of disorder (726, 735-736). They then created a model for predicting gentrification as measured by GGO score using these other characterisitic.

In this research, the authors classified gentrification at the block-face level as follows.
1. "Structural mix": condition of older structures and degree of rehabilitation and new structures, measured by the following.
    - Number of new or rehabbed buildings.
    - New traffic structures or signs.
    - New large developments.
    - New construction for sale.
    - New public courtesies. (733)
2. "Visible beautification efforts", measured as follows:
    - Painting over graffiti.
    - Grooming personal frontage.
    - Beautifying vacant/public space. (733)
3. "Lack of disorder and decay", measured by the lack of the following:
    - Unkempt vacant/public space.
    - Physical disorder.
    - Decaying structures. (733)

The researchers then calculated a summary score for each of these three measures and averaged the three measures for a final score (733). 

The authors argue that it is important to use visual cues to detect gentrification, because previous studies that did not (e.g. ones that use census data only, as surveyed by Wyly and Hammel 1996, 1998, 1999) classified many neighborhoods incorrectly (731). Thus, this research is important for improving our understanding on gentrification.

### Human Computation
