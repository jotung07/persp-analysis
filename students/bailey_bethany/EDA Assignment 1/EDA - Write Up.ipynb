```python
# Import libraries and dataframe
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
df = pd.read_csv("gss2012.csv")
```

## Missing Populations
Throughout my exploratory data analysis (EDA) of the General Social Survey data, I was interested in looking at the breakdowns of variables across race, sex, political party, religion, and age. I expected that it would be difficult to accurately complete these analyses because the categories would not be equally weighted across the sample population. In order to prevent simple mistakes in my EDA, I completed an initial analysis of the data demographics. Through this analysis, I realized that there was an unequal distribution of males and females in the data, so I would need to keep that in mind or adjust for survey weights when I compared between sexes. There were also unequal weightings of age distribution (the data skewed lower and had fewer elderly people), political party (there were fewer conservatives than democrats), religion (the majority of individuals were Christian/Catholic), and race (there fewer African Americans and the "Other" group was not broken out). For example, below are the breakdowns for sex and age, which show the previously mentioned skews in these data.

Thus, comparing absolute weights between groups was difficult without weighting the data. I found my most productive and interesting insights by comparing relative weights within groups.

```python
sns.countplot(x="sex", data=df)
plt.title('Sex Breakdown')
plt.xlabel('Sex')
plt.ylabel('Count')
```

```python
pd.DataFrame.hist(df, column="age", bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
```

## Interesting Questions
After discovering and noting the skewed demographics, I found some interesting questions I would like to explore with further analysis.

### Question 1: How and why do voting rates change?
Looking at 2008 voting rates among individuals in different political parties, I found that independents were the only group that did not vote more than they voted in 2008. I theorized that this might be correlated to another varaible, such as confidence in government. Thus, I plotted confidence in government for voters and non-voters. Surprisingly, I found that voters were less confident in government than non-voters, which was counter to my hypothesis. Then, I wanted to see whether this confidence in government varied greatly between political parties, in order to see whether there was a notable effect on whether or not people voted. I did not see any major correlation. However, though this analysis was not effective, I hypothesize that there are other interesting trends and correlations among voting and non-voting populations.

Potential research questions I might answer are as follows: 

1. How do independents differ from other political parties across different variables? 
2. Are there significant correlations between voting likelihood and specific political beliefs? Do these correlations change across different election cycles? 

```python
p = sns.factorplot(x="voted08", col="partyid", kind="count", data=df)
p.set(xlabel='Voted in 2008', ylabel='Count')
plt.subplots_adjust(top=.8)
p.fig.suptitle('Voting in 2008 by Party')
```

```python
sns.boxplot(x="voted08", y="con_govt", data=df)
plt.title("Confidence in Government among Voting and Non-Voting Population")
plt.xlabel("Voted in 2008")
plt.ylabel("Confidence in Government")
```

### Question 2: How do beliefs change with education?
After looking at the different questions asked in the survey, I was interested in seeing what influences peoples' answers to the question about whether we should allow anti-American Muslim clergymen to teach in college. I found slight differences in this belief across political parties (Independent Democrats were almost equally likely to say yes, while all other parties were very likely to say no). Also, the people who answered yes were skewed towards middle-aged individuals. Perhaps the most interesting finding was that the ratio of individuals who said no decreased as educational attainment increased (see below graph). That is, a smaller portion of people with higher educational attainment believed that we should not allow anti-American Muslim clergymen to teach in college.If given a more complete dataset, I would be interested in asking the following:

1. Does this trend hold with a larger sample size?
2. How has this trend varied over time? Were less educated people more likely to answer yes to this question before early 2000 (especially given the events of 9/11)? Has this difference in attitudes across levels of educational attainment always existed?

Additionally, given this finding, I would look for similar or opposite patterns in other variables as educational attainment changes. I hypothesize that, using this data, we could find many such changes in beliefs and actions as people become more educated. Some potential questions in this format are as follows:

1. Adjusted for age, do people tend to have fewer children as they become more educated?
2. How do people's views about female roles change as they become more educated? What about their views on abortion?
3. Are people more or less likely to think positively about autoritarianism as they become more educated?

```python
g = sns.factorplot(x="colmslm", col="degree", col_order=("<HS", "HS", "Junior Coll", "Bachelor deg", \
                            "Graduate deg"), kind="count", data=df)
g.set(xlabel='Allowed to teach? y/n', ylabel='Count')
plt.subplots_adjust(top=.8)
g.fig.suptitle('Answers to "Should Anti-American Muslim Clergymen be allowed to teach?" by College Attainment')
```
